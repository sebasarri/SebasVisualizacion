```python
df3 = pd.read_csv('datasets/cwurData.csv')
df3.head(2)
```

```python
def create_category(ranking): 
    if(ranking >=1)&(ranking<=100):
       return 'First tier university'
    elif(ranking >=100)&(ranking<=200):
       return 'Second tier university'
    elif(ranking >=200)&(ranking<=300):
       return 'Third tier university'
    else:
       return 'Other top university'
```

```python
df3['Rank_level'] = df3['world_rank'].apply(lambda x:create_category(x))
```
